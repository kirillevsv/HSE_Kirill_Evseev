```python
seconds_request = input("Введите количество секунд: ")
if str(seconds_request).isdigit():
  seconds_request = int(seconds_request)
  if seconds_request < 86400:
    hours = int(seconds_request // 3600)
    minutes = int((seconds_request - (hours * 3600)) // 60)
    seconds = int((seconds_request - (hours * 3600)) - (minutes * 60))
    print(f" Текущее время - {hours} ч. {minutes} мин. {seconds} сек")
  elif seconds_request == 86400:
    print(f" Текущее время - 00:00:00")
  else:
    print("Ошибка - количество секунд не более 86 400")
else:
  print("Ошибка, введите только цифры")
```
