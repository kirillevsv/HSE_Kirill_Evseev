```python
# Расчет факториала числа
def fact(n):
  res = 1
  for i in range(1, n+1):
    res = res * i
  return res
print (fact(5))
```

```python
# Вычисление наибольшего числа
def max_three(nums):
    a, b, c = nums
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
print(max_three((1, 2, 3)))
```
